```python
#changing to Nan
cname = ['temp','atemp', 'hum','windspeed','casual','registered','cnt']
numeric_data = data[cname]
         
for i in cname:
    print(i)
    q75, q25 = np.percentile(data.loc[:,i],[75,25])
    iqr = q75- q25
    min = q25 - (1.5*iqr)
    max = q75 + (1.5*iqr)
    
    data.loc[data.loc[:,i] < min , i] = np.nan
    data.loc[data.loc[:,i] > max , i] = np.nan   
    
#Deleting purposfully so that while imputing we can chek the results based on these values as we already kbow the values here
#data.iloc[150,11] = np.nan  
#data.iloc[150,12] = np.nan  
#data.iloc[150,13] = np.nan
```

```python
#Imputing Missing value
#1. Imputing hum      

#Actual data.iloc[150,11] = 0.636667
data.iloc[150,11] = np.nan 
#data['hum'] = data['hum'].fillna(data['hum'].median()) #=0.6278275
#data['hum'] = pd.DataFrame(KNN(k=3).fit_transform(data[['hum']])) #= 0.0
data['hum'] = data['hum'].fillna(data['hum'].mean()) # = 0.6293488

print(data.iloc[150,11]) 
# so we are imputing this value with mean since that is the closest value
```

```python
#2. Imputing windspeed      

#Actual data.iloc[150,12] = 0.111329
data.iloc[150,12] = np.nan 
data['windspeed'] = data['windspeed'].fillna(data['windspeed'].median()) #=0.179108 
#data['windspeed'] = pd.DataFrame(KNN(k=3).fit_transform(data[['windspeed']])) #= 0.0
#data['windspeed'] = data['windspeed'].fillna(data['windspeed'].mean()) # = 0.18636491876712305

print(data.iloc[150,12]) 
# so we are imputing this value with median since that is the closest value
```

```python
#3. imputing missing casual
#original data.iloc[150,13] = 673
data.iloc[150,13] = np.nan 
#data['casual'] = data['casual'].fillna(data['casual'].median()) # 675
#data['casual'] = data['casual'].fillna(data['casual'].mean()) #732
#data['casual'] = pd.DataFrame(KNN(k=3).fit_transform(data[['casual']])) #= 0.0
data['casual'] = data['casual'].fillna(data['cnt']-data['registered']) # we saw that this is the exact formula to caclulate the casual variable
print(data.iloc[150,13]) 
```

```python
#********************************** Modeling ************************************
##Sampling: dividing  Test and train data  using sklearn 
from sklearn.model_selection import train_test_split,KFold, cross_val_score, cross_val_predict

#Random sample selection 
train, test = train_test_split(data, test_size = 0.20, random_state = 100)
train.to_csv("TrainFile_BikeRenting_python.csv", index=False)
test.to_csv("TestFile_BikeRenting_python.csv", index=False)
data.shape, test.shape , train.shape
```

```python
from sklearn import metrics
#function to check performance  
def performance(actual, predict):
    print('MSE:', metrics.mean_squared_error(actual, predict)) 
    print('RMSE:', np.sqrt(metrics.mean_squared_error(actual, predict))) 
    print('MAPE:',np.mean(np.abs((actual-predict)/actual))*100)
    print('R-Sq:', metrics.r2_score(actual, predict))
```

```python
#1. ############################################ Decision Tree ############################
from sklearn.tree import DecisionTreeRegressor

dt1 = DecisionTreeRegressor(max_depth =6, random_state=123).fit(train.iloc[:,0:7],train.iloc[:,7])
prediction_dt1 = dt1.predict(test.iloc[:,0:7])

#error matrix
performance(test.iloc[:,7],prediction_dt1)
print('                           ')
print('Perdicted Vs Actual value: ')
prediction_dt1[9], test.iloc[9,7]
```

```python
#2. ############################################ Random Forest ########################################################
from sklearn.ensemble import RandomForestRegressor

rf1 = RandomForestRegressor(n_estimators = 100, random_state = 123).fit(train.iloc[:,0:7],train.iloc[:,7])
prediction_rf1 = rf1.predict(test.iloc[:,0:7])

performance(test.iloc[:,7],prediction_rf1)
print('                           ')
print('Perdicted Vs Actual value: ')
prediction_rf1[1], test.iloc[1,7]
```

```python
#3. ############################################ Linear Regression ############################
from sklearn.linear_model import LinearRegression

ln_model = LinearRegression().fit(train.iloc[:,0:7],train.iloc[:,7])

prediction_lr1 = ln_model.predict(test.iloc[:,0:7])

performance(test.iloc[:,7],prediction_lr1)
print('                           ')
print('Perdicted Vs Actual value: ')
prediction_lr1[1], test.iloc[1,7]
```

```python
#Ploting to understand the spread of predicted data. 

plt.figure(figsize=(14,4))

plt.subplot(1,3,1)
plt.title('Decision Tree predictions')
plt.scatter(test.iloc[:,7] , prediction_dt1)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')

plt.subplot(1,3,2)
plt.title('Random Forest predictions')
plt.scatter(test.iloc[:,7] , prediction_rf1)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')

plt.subplot(1,3,3)
plt.title('Linear Regression Predictions')
plt.scatter(test.iloc[:,7] , prediction_lr1)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')

plt.tight_layout()
```
